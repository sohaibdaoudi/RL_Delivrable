## 1. My Framework v1 (ispired by gymnasioum)

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import time
```

```python
class GridWorld:
    def __init__(self, height = 5, width = 5, action_space_size = 4, action2direction = {
            0: np.array([-1,0]), # up
            1: np.array([1,0]), # down
            2: np.array([0,-1]), # left
            3: np.array([0,1]) # right
        }, target_location = (0,0),  agent_location = (4,4)):
        self.size = (height, width)
        self.action_space_size = action_space_size

        self.action2direction = action2direction
        
        self.target_location = target_location
        self.agent_location= agent_location
        self.reset()


    def get_size(self):
        return self.size
    def set_size(self, h, w):
        self.size = (h,w)
        
    def get_action_space_size(self):
        return self.action_space_size
    def set_action_space_size(self, n):
        self.action_space_size = n
        
    def get_action2direction(self):
        return self.action2direction
    def set_action2direction(self, mapping):
        self.action2direction = mapping
        
    def get_target_location(self):
        return self.target_location
    def set_target_location(self, tupl):
        self.target_location = tupl
        
    def get_agent_location(self):
        return self.agent_location
    def set_agent_location(self, tupl):
        self.agent_location = tupl



    
    def reset(self):
        size = self.get_size()
        a = (np.random.randint(0, size[0]), np.random.randint(0, size[1]))
        self.set_target_location(a)
        while True:
            b = (np.random.randint(0, size[0]), np.random.randint(0, size[1]))
            if b != self.get_target_location():
                break

        self.set_agent_location(b)


    def step(self, action):

        agent_position = np.array(self.get_agent_location())
        
        mapping = self.get_action2direction()
        direction = mapping[action]
    
        next_position = agent_position + direction
        size = self.get_size()

        if ((0 <= next_position[0] < size[0]) and (0 <= next_position[1] < size[1])):
            self.set_agent_location(tuple(next_position))

        target = self.get_target_location()
        current_position = self.get_agent_location()
        finished = (current_position == target)
        reward = 10 if finished else -1

        info = {}

        return current_position, reward, finished, info


    def render(self):
        os.system('cls' if os.name == 'nt' else 'clear')

        print(f"Agent: {self.get_agent_location()} | Target: {self.get_target_location()}\n")

        grid = np.full(self.get_size(), 'x')
        grid[self.get_agent_location()] = 'A'
        grid[self.get_target_location()] = 'T'
        

        for row in grid:
            print(' '.join(row))
        print("\n")
                   
```

### a. Test without customization

```python
env = GridWorld()
state = env.reset()
env.render()

finished = False
total_reward = 0
max_steps = 25

for step in range(1, max_steps+1):
    action = np.random.randint(0, env.get_action_space_size())


    new_state, reward, finished, info = env.step(action)
    
    total_reward+=reward

    env.render()
    print(f"Cumulative Reward = {total_reward}")
    print("\n" + "="*10)
    
    time.sleep(0.3)



    
    if finished:
            print(f"\nTarget reached in {step} steps! Total Reward : {total_reward}")
            break   



if not finished:
    print("\nAgent did not reach the target within the step limit.")
```

## 2. My Framework v2 with plots (ispired by gymnasioum)

```python
import matplotlib.pyplot as plt
import numpy as np
import time
from matplotlib.animation import FuncAnimation





class GridWorld:
    def __init__(self, height = 5, width = 5, action_space_size = 5, action2direction = {
            0: np.array([-1,0]), # up
            1: np.array([1,0]), # down
            2: np.array([0,-1]), # left
            3: np.array([0,1]), # right
            4: np.array([0,0]) # stay
        }, target_location = (0,0),  agent_location = (4,4)):
        self.size = (height, width)
        self.action_space_size = action_space_size

        self.action2direction = action2direction
        
        self.target_location = target_location
        self.agent_location= agent_location
        self.history = [] # to track the agent
        self.reset()


    def get_size(self):
        return self.size
    def set_size(self, h, w):
        self.size = (h,w)
        
    def get_action_space_size(self):
        return self.action_space_size
    def set_action_space_size(self, n):
        self.action_space_size = n
        
    def get_action2direction(self):
        return self.action2direction
    def set_action2direction(self, mapping):
        self.action2direction = mapping
        
    def get_target_location(self):
        return self.target_location
    def set_target_location(self, tupl):
        self.target_location = tupl
        
    def get_agent_location(self):
        return self.agent_location
    def set_agent_location(self, tupl):
        self.agent_location = tupl

    def get_history(self):
        return self.history
    def set_history(self, Dic):
        self.history = Dic



    
    def reset(self):
        size = self.get_size()
        a = (np.random.randint(0, size[0]), np.random.randint(0, size[1]))
        self.set_target_location(a)
        while True:
            b = (np.random.randint(0, size[0]), np.random.randint(0, size[1]))
            if b != self.get_target_location():
                break

        self.set_agent_location(b)
        self.history = [{'agent': self.get_agent_location(), 'target': self.get_target_location(), 'reward': 0}]
        

    def step(self, action):

        agent_position = np.array(self.get_agent_location())
        
        mapping = self.get_action2direction()
        direction = mapping[action]
    
        next_position = agent_position + direction
        size = self.get_size()

        if ((0 <= next_position[0] < size[0]) and (0 <= next_position[1] < size[1])):
            self.set_agent_location(tuple(next_position))
        else:
            self.set_agent_location(tuple(agent_position))


        target = self.get_target_location()
        current_position = self.get_agent_location()
        finished = (current_position == target)
        reward = 10 if finished else -1

        self.history.append({'agent': current_position, 'target': target, 'reward': reward})
        
        return current_position, reward, finished



    def Create_Visualization(self):
        fig, ax = plt.subplots(figsize=(7, 8))
        fig.subplots_adjust(bottom=0.15)
    
        def setup_grid():
            ax.set_xlim(-0.5, self.size[1] - 0.5)
            ax.set_ylim(-0.5, self.size[0] - 0.5)
    
            ax.set_xticks(np.arange(-0.5, self.size[1], 1), minor=True)
            ax.set_yticks(np.arange(-0.5, self.size[0], 1), minor=True)
            ax.grid(which='minor', color='#CCCCCC', linestyle='-', linewidth=1)
            ax.tick_params(which='minor', bottom=False, left=False, labelbottom=False, labelleft=False)
            ax.set_aspect('equal', adjustable='box')
            ax.invert_yaxis()
    
        setup_grid()
    
        # target drawing
        target_pos = self.history[0]['target']
        ax.scatter(target_pos[1], target_pos[0], marker=(5, 1, 0), s=800, c='#FFD700',
                   edgecolors='black', zorder=5, label='Target')
    
        # agent + path drawing
        agent_marker = ax.scatter([], [], s=400, c='#0077BE', edgecolors='white',
                                  linewidth=2, zorder=5, label='Agent')
        path_line, = ax.plot([], [], color='#0077BE', alpha=0.5, linewidth=4, zorder=4)
    
        ax.legend(loc='upper right')
    
        info_text = fig.text(0.5, 0.05, '', ha='center', fontsize=16, fontweight='bold',
                             bbox=dict(boxstyle='round,pad=0.5', fc='lightgray', alpha=0.5))
    
        cumulative_rewards = np.cumsum([frame['reward'] for frame in self.history])
    
        def update(frame_num):
            current_frame = self.history[frame_num]
            agent_pos = current_frame['agent']
    
            # Update agent
            agent_marker.set_offsets([agent_pos[1], agent_pos[0]])
    
            # Update path
            path_rows, path_cols = zip(*[f['agent'] for f in self.history[:frame_num + 1]])
            path_line.set_data(path_cols, path_rows)
    
            # Title and info text
            ax.set_title(f"Agent at: {agent_pos}", fontsize=14)
            total_reward_so_far = cumulative_rewards[frame_num]
            info_text.set_text(f"Step: {frame_num:02d}   |   Cumulative Reward: {total_reward_so_far}")
    
            if frame_num == len(self.history) - 1 and agent_pos == target_pos:
                ax.set_title("Target Reached!", fontsize=16, color='green')
    
            return agent_marker, path_line, info_text
    
        anim = FuncAnimation(fig, update, frames=len(self.history), interval=250, blit=False)
    
        plt.close(fig)
        return anim


```

```python
env = GridWorld(height = 5, width = 5, action_space_size = 5, action2direction = {
            0: np.array([-1,0]), # up
            1: np.array([1,0]), # down
            2: np.array([0,-1]), # left
            3: np.array([0,1]), # right
            4: np.array([0,0]) # stay
        })

state = env.reset()

finished = False
max_steps = 25

for step in range(1, max_steps + 1):
    action = np.random.randint(0, env.get_action_space_size())
    new_state, reward, finished = env.step(action)
    if finished:
        print(f"Target reached in {step} steps!")
        break

if not finished:
    print(f"Agent did not reach the target within {max_steps} steps.")

# --- create and Display the Animation ---
anim = env.Create_Visualization()
HTML(anim.to_html5_video())
```
